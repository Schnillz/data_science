```python
%matplotlib inline
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
# http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html
iris = datasets.load_iris()
list(iris.keys())
```

### Getting number of occurences taking the boundaries of exercise 9  
### Question again: is flower of type virginica?

```python
X_sepal_length = iris["data"][:, :1]  # sepal length
X_sepal_width = iris["data"][:, 1:2]  # sepal width
X_petal_length = iris["data"][:, 2:3]  # sepal width
X_petal_width = iris["data"][:, 3:]  # sepal width

setosa = (iris["target"] == 0).astype(np.int)  # array of 1 if Iris-Setosa, else 0
versicolor = (iris["target"] == 1).astype(np.int)  # array of 1 if Iris-Versicolor, else 0
virginica = (iris["target"] == 2).astype(np.int)  # array of 1 if Iris-Virginica, else 0

boundary_sepal_length = np.where(X_sepal_length > 6.9)[0]
boundary_petal_length = np.where(X_petal_length > 4.7)[0]
boundary_petal_width = np.where(X_petal_width > 1.6)[0]

indices_hits_0_sepal_length = np.take(setosa, boundary_sepal_length)
indices_hits_1_sepal_length = np.take(versicolor, boundary_sepal_length)
indices_hits_2_sepal_length = np.take(virginica, boundary_sepal_length)
hits_0_sepal_length = np.count_nonzero(indices_hits_0_sepal_length) + 1
hits_1_sepal_length = np.count_nonzero(indices_hits_1_sepal_length) + 1
hits_2_sepal_length = np.count_nonzero(indices_hits_2_sepal_length) + 1
print('Hits of Setosa with sepal length > 6.9 :: ' + str(hits_0_sepal_length))
print('Hits of Versicolor with sepal length > 6.9 :: ' + str(hits_1_sepal_length))
print('Hits of Virginica with sepal length > 6.9 :: ' + str(hits_2_sepal_length))
print('\n')

indices_hits_0_petal_length = np.take(setosa, boundary_petal_length)
indices_hits_1_petal_length = np.take(versicolor, boundary_petal_length)
indices_hits_2_petal_length = np.take(virginica, boundary_petal_length)
hits_0_petal_length = np.count_nonzero(indices_hits_0_petal_length) + 1
hits_1_petal_length = np.count_nonzero(indices_hits_1_petal_length) + 1
hits_2_petal_length = np.count_nonzero(indices_hits_2_petal_length) + 1
print('Hits of Setosa with petal_length > 4.7 :: ' + str(hits_0_petal_length))
print('Hits of Versicolor with petal_length > 4.7 :: ' + str(hits_1_petal_length))
print('Hits of Virginica with petal_length > 4.7 :: ' + str(hits_2_petal_length))
print('\n')

indices_hits_0_petal_width = np.take(setosa, boundary_petal_width)
indices_hits_1_petal_width = np.take(versicolor, boundary_petal_width)
indices_hits_2_petal_width = np.take(virginica, boundary_petal_width)
hits_0_petal_width = np.count_nonzero(indices_hits_0_petal_width) + 1
hits_1_petal_width = np.count_nonzero(indices_hits_1_petal_width) + 1
hits_2_petal_width = np.count_nonzero(indices_hits_2_petal_width) + 1
print('Hits of Setosa with petal_width > 1.6 :: ' + str(hits_0_petal_width))
print('Hits of Versicolor with petal_width > 1.6 :: ' + str(hits_1_petal_width))
print('Hits of Virginica with petal_width > 1.6 :: ' + str(hits_2_petal_width))
```

### B)
Check your tree against the tree derived from SciKit Learn as given in the Python example before! Print the tree with Graphviz (can be easily done with WebGraphViz [2]) 

Because of encoding issues done with sepal length and sepal width

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from IPython.display import Image

X = iris.data[:, :2] # sepal length and sepal width
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)

export_graphviz(
        tree_clf,
    out_file="./iris_tree.dot",
        feature_names=iris.feature_names[:2],
        class_names=iris.target_names,
        rounded=True,
        filled=True
)

Image("iris_tree.png", width=640, height=640)
```
